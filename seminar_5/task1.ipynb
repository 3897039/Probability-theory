Проведите тест гипотезы. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, 
имеют средний диаметр 17 мм. Используя односторонний критерий с α=0,05, проверить эту гипотезу, 
если в выборке из n=100 шариков средний диаметр оказался равным 17.5 мм, 
а дисперсия известна и равна 4 кв. мм.

Наблюдаемое значение критерия U равно:
U = (17.5 - 17)*sqrt(100) = 5

Уровень значимости а = 0,05
Ф(U) = (1 - 2a)/2 = 0.45
критическая точка u = 1.645

Т.к. U < u, то нулевую гипотезу можно принять, можно считать что средний диаметр действительно равен 17 см.
